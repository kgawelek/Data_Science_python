### practising merge() options

```python
suppliers = pd.read_csv('../course-files/northwind-mongo-master/suppliers _del_1.csv', usecols=['SupplierID', 'CompanyName', 'City'])
products_del2 = pd.read_csv('../course-files/northwind-mongo-master/products_del_2.csv', usecols=['ProductID','ProductName', 'SupplierID'])
```

```python
prod_and_sup = suppliers.merge(products_del2, how='inner', on='SupplierID', indicator=True)
```

```python
print(len(suppliers))
print(len(products_del2))
print(len(prod_and_sup))
```

```python
sup_prod_left = suppliers.merge(products_del2, how='left', on='SupplierID', indicator=True)
```

```python
print(len(suppliers))
print(len(products_del2))
print(len(sup_prod_left))
```

```python
sup_prod_left.query('_merge == "left_only"')
```

```python
products = pd.read_csv('../course-files/northwind-mongo-master/products.csv', usecols=['ProductID', 'ProductName', 'SupplierID'])
```

```python
suppliers.merge(products, on='SupplierID').head()
```

```python
suppliers.set_index('SupplierID', inplace=True)
```

```python
suppliers.merge(products, left_index=True, right_on='SupplierID')
```

```python
products.set_index('SupplierID', inplace=True)
```

```python
suppliers.merge(products, left_index=True, right_index=True).head()
```

```python
suppliers.reset_index(inplace=True)
```

```python
suppliers.merge(products, left_on='SupplierID', right_index=True).head()
```

```python
products.reset_index(inplace=True)
products.rename({'SupplierID': 'supplier_id'}, axis='columns', inplace=True)
```

```python
suppliers.merge(products, left_on='SupplierID', right_on='supplier_id').head()
```

## Merging more than 2 data frames

```python
prod_cat = products.merge(categories, left_on='CategoryID', right_index=True)
```

```python
prod_cat.head()
```

```python
prod_cat_sup = prod_cat.merge(suppliers, on='SupplierID')
```

```python
prod_cat_sup.head()
```

```python
products.merge(categories, left_on='CategoryID', right_index=True).merge(suppliers, on='SupplierID').head()
```

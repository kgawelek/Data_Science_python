```python
fuel = pd.read_csv('../course-files/fuel.csv', low_memory=False, usecols=[ 'Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)','Start Stop Technology'], index_col='Vehicle ID')
```

```python
list_of_cars = ['Renault', 'Toyota', 'Ford']
```

```python
is_in_list = fuel['Make'].isin(list_of_cars)
```

```python
fuel[is_in_list].head()
```

```python
is_start_stop_defined = fuel['Start Stop Technology'].notnull()
```

```python
fuel[is_start_stop_defined].tail()
```

```python
start_stop_cars = fuel[fuel['Start Stop Technology'] == 'Y']
```

```python
start_stop_cars.head()
```

```python
is_not_start_stop_def = fuel['Start Stop Technology'].isnull()
```

```python
fuel[is_not_start_stop_def].head()
```

```python
mpg50_60 = fuel['Combined MPG (FT1)'].between(50, 60)
```

```python
mpg50_60.head()
```

```python
fuel[mpg50_60].head()
```

```python
print('There are ', len(fuel[is_in_list]), ' cars in ', list_of_cars)
print('There are ', len(fuel[is_start_stop_defined]), ' cars where Start/Stop is defined')
print('There are ', len(fuel[mpg50_60]), ' cars with MPG between 50 and 60')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
education = pd.read_csv("../course-files/StackOverflowDeveloperSurvey2018.csv", usecols=['FormalEducation', 'Respondent'], squeeze=True, index_col='Respondent')
```

```python
education.head()
```

```python
education.value_counts()
```

```python
education.where(education == 'Doctoral degree').dropna()
```

```python
education.where(education == "Bachelor's degree").dropna()
```

```python
bachelors = education.where(education == "Bachelor’s degree (BA, BS, B.Eng., etc.)").dropna()
doctors = education.where(education == 'Other doctoral degree (Ph.D, Ed.D., etc.)').dropna()
masters = education.where(education == "Master’s degree (MA, MS, M.Eng., MBA, etc.)").dropna()
```

```python
doctors.size
```

```python
degrees = pd.Series({
    "doctors": len(doctors),
    "bachelors": len(bachelors),
    "masters": len(masters)
}, name='degrees percantage')
```

```python
degrees.plot.pie()
plt.show()
```

```python
degrees.plot.bar()
plt.show()
```

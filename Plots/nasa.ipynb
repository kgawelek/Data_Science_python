```python
nasa = pd.read_csv("../course-files/nasa_facebook_statuses.csv",usecols=["status_published","num_reactions","num_comments","num_shares","num_likes"])
nasa["status_published"] = pd.to_datetime(nasa["status_published"])
filter =(nasa['status_published'] >= '2016-3-1') & (nasa['status_published'] < '2016-6-1')
nasa_march_may = nasa[filter]
nasa_march_may['month'] = nasa_march_may.apply(lambda row: row["status_published"].month,axis=1)
nasa_by_month = nasa_march_may.groupby(by='month').sum()
nasa_by_month.head()
```

```python
nasa_by_month.rename(index={3: 'MARCH', 4: 'APRIL', 5: 'MAY'}, inplace=True)
```

```python
nasa_by_month
```

```python
plt.style.use('default')
nasa_by_month['num_likes'].plot(kind='pie', figsize=(5, 5))
```

```python
nasa_by_month['num_likes'].plot(kind='pie', figsize=(5, 5), title='NASA Facebook activity', labels=('III', 'IV', 'V'))
```

```python
nasa_by_month.plot(kind='pie', subplots=True, figsize=(10, 10), legend=False, layout=(2,2))
```

```python
nasa_by_month.plot(kind='pie', subplots=True, figsize=(10, 10), legend=True, layout=(2,2), autopct='%.0f%%', textprops={'color': 'w'})
```
